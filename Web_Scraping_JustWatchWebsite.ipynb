## **Fetching Movie URL's**

````python
# Write Your Code here
import random  # Importing the random module to help shuffle and randomly select links

# STEP 1C: Randomly fetch 100 movie links from multiple pages on JustWatch

movie_links = []  # Initialize an empty list to store unique movie URLs

# Loop through the first 10 pages of the JustWatch movie listings
for page_num in range(1, 11):  # Loop from page 1 to 10 (inclusive) to get a broad range of movie links
    # Construct the URL for the current page number
    url = f"https://www.justwatch.com/in/movies?release_year_from=2000&page={page_num}"

    # Use Selenium to load the page (this ensures JavaScript-rendered content is also loaded)
    driver.get(url)
    time.sleep(3)  # Wait for 3 seconds to allow the page content to fully load

    # Parse the loaded HTML content using BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Select all anchor tags that link to individual movie pages
    anchors = soup.select('a.title-list-grid__item--link')

    # Loop through each anchor tag and extract its full URL
    for a in anchors:
        full_link = "https://www.justwatch.com" + a.get('href')  # Form the complete URL
        if full_link not in movie_links:  # Avoid duplicates
            movie_links.append(full_link)  # Add the unique movie URL to the list

# Shuffle the collected links to ensure randomness
random.shuffle(movie_links)

# Retain only the first 100 movie links after shuffling
movie_links = movie_links[:100]

# Print summary and a small sample of the collected links
print("Random Movie URLs Collected:", len(movie_links))  # Should print 100 if successful
print(movie_links[:5])  # Preview the first 5 links to verify the format

````

````python
# STEP 2A: Initialize empty lists to store data scraped from each movie page

movie_titles = []          # To store the title of each movie
release_years = []         # To store the release year
genres = []                # To store the genre(s) (e.g., Action, Drama)
imdb_ratings = []          # To store IMDb ratings (if available)
runtimes = []              # To store runtime (e.g., 1h 45m)
age_ratings = []           # To store age ratings (e.g., U/A 13+, PG-13)
countries = []             # To store country of origin (e.g., USA, India)
streaming_services = []    # To store streaming platforms (e.g., Netflix, Prime Video)
movie_urls = []            # To keep track of the successfully scraped URLs

# Loop through the list of collected movie links (from the previous step)
for idx, link in enumerate(movie_links):
    print(f"Scraping {idx+1}/{len(movie_links)}: {link}")  # Print progress info for each movie

    try:
        # Use Selenium to load the movie detail page
        driver.get(link)
        time.sleep(3)  # Wait for 3 seconds to allow JavaScript to load the content

        # Parse the page using BeautifulSoup
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        # Store the current movie URL (for reference or export later)
        movie_urls.append(link)

        # Optional: Save the parsed HTML (soup) of each movie page as a global variable
        # This allows you to access it later using variables like soup_movie_0, soup_movie_1, etc.
        globals()[f"soup_movie_{idx}"] = soup

        # Note: Extraction of specific data (title, rating, etc.) will be handled in the next steps/cells.

    except Exception as e:
        print("Error scraping:", e)  # Print the error for debugging purposes

        # In case of failure (page didn't load, error occurred), append None values to maintain list alignment
        movie_titles.append(None)
        release_years.append(None)
        genres.append(None)
        imdb_ratings.append(None)
        runtimes.append(None)
        age_ratings.append(None)
        countries.append(None)
        streaming_services.append(None)
        movie_urls.append(link)  # Still record the link that failed for potential reprocessing later
````

## **Scrapping Movie Title**

````python
# Write Your Code here
# STEP 2B: Extract Movie Titles from the previously stored soup objects

# Loop over the range of movie_links to match index with soup_movie_0, soup_movie_1, etc.
for idx in range(len(movie_links)):
    try:
        # Dynamically retrieve the BeautifulSoup object for the current movie using globals()
        # This accesses variables like soup_movie_0, soup_movie_1, ... stored earlier
        soup = globals()[f"soup_movie_{idx}"]

        # Try to select the main movie title:
        # First, it looks for an <h1> tag with a specific data-testid attribute used by JustWatch
        # If that doesn't exist, it falls back to selecting any <h1> tag (as a backup)
        title_tag = soup.select_one('h1[data-testid="title"]') or soup.select_one('h1')

        # Extract and clean the text inside the tag (if tag was found), else set title to None
        title = title_tag.text.strip() if title_tag else None

        # Append the extracted title to the movie_titles list
        movie_titles.append(title)

    except Exception as e:
        # If there's any error (e.g., soup not found, tag missing), add None to keep list aligned
        movie_titles.append(None)

        # Print error message along with the index for easy debugging
        print(f"Title error at index {idx} →", e)

# Display the first 5 extracted titles to verify that the scraping worked
print("Extracted Titles:", movie_titles[:5])
````

## **Scrapping release Year**

````python
# Write Your Code here
# STEP 2C: Extract Release Year from soup_movie_* objects using regex
import re  # Import Python's regular expression module to search for year patterns like (2023)

# Loop through all movie indices to access their corresponding soup objects
for idx in range(len(movie_links)):
    try:
        # Retrieve the corresponding BeautifulSoup object stored earlier
        soup = globals()[f"soup_movie_{idx}"]

        # First attempt: Try to extract release year from the <h1> movie title header
        # Example: "Inception (2010)" → extract (2010)
        header = soup.find("h1")  # Get the main heading tag

        if header:
            # Use regular expression to find a 4-digit year inside parentheses
            match = re.search(r"\((\d{4})\)", header.text)
            if match:
                release_years.append(match.group(1))  # Extract and append the year (e.g., "2010")
                continue  # Skip the fallback method if year is already found

        # Fallback method: search for a year pattern anywhere else in the page content
        # This captures cases where the year is not in the <h1> title
        year_tag = soup.find(string=re.compile(r'\b(20\d{2}|19\d{2})\b'))  # Search for 4-digit year in page text

        # Extract the actual year if found, else set as None
        year = re.search(r'\b(20\d{2}|19\d{2})\b', year_tag).group(0) if year_tag else None
        release_years.append(year)  # Append the result to the release_years list

    except Exception as e:
        # If an error occurs (e.g., tag missing or regex fails), append None to keep the list aligned
        release_years.append(None)
        print(f"Release year error at index {idx} →", e)  # Print debug message for tracking issues

# Preview the first 5 extracted release years to verify correctness
print("Extracted Release Years:", release_years[:5])
````

## **Scrapping Genres**

````python
# Write Your Code here
# STEP 2D: Final version to extract genres from soup_movie_* using the correct h3-based HTML structure
genres = []  # Initialize an empty list to store genres for each movie

# Loop through each index corresponding to the soup_movie_* objects
for idx in range(len(movie_links)):
    try:
        # Retrieve the stored BeautifulSoup object for the current movie
        soup = globals()[f"soup_movie_{idx}"]

        genre = 'N/A'  # Default value in case genre is not found

        # Look for all <h3> tags with the specific class used to label metadata sections
        # This is where section labels like "Genre", "Age Rating", "Runtime", etc., appear
        h3_tags = soup.find_all('h3', class_='poster-detail-infos__subheading')

        # Loop through all found <h3> tags
        for h3 in h3_tags:
            # Check if the text inside the <h3> contains the word "Genre" (covers both "Genre" and "Genres")
            if 'Genre' in h3.get_text(strip=True):

                # The genre values are usually inside a <div> immediately after the <h3> tag
                genre_div = h3.find_next_sibling('div', class_='poster-detail-infos__value')

                if genre_div:
                    # Find all tags inside that <div> that hold individual genre labels (usually <span> or <a>)
                    genre_spans = genre_div.find_all(['span', 'a'])

                    # Extract the cleaned text from each tag and remove any trailing commas
                    genre_list = [
                        g.get_text(strip=True).strip(',')
                        for g in genre_spans
                        if g.get_text(strip=True)
                    ]

                    # If genres are found, remove duplicates, sort them alphabetically, and join with commas
                    genre = ", ".join(sorted(set(genre_list))) if genre_list else 'N/A'
                break  # Exit the loop after successfully finding and processing the Genre section

        # Append the final genre string (or 'N/A') to the genres list
        genres.append(genre)

    except Exception as e:
        # If any error occurs (e.g., tags not found, parsing issue), store 'N/A' as placeholder
        genres.append('N/A')
        print(f"Genre extraction error at index {idx} →", e)  # Print the error for debugging

# Show the first 5 extracted genres to verify success
print("Extracted Genres (Reliable):", genres[:5])
````

## **Scrapping IMBD Rating**

````python
# Write Your Code here
# STEP 2E (Final): Extract IMDb ratings by locating the 'IMDB' logo and its nearby text
import re  # Import the regular expressions module to help extract rating numbers like 7.8
imdb_ratings = []  # Initialize a list to store IMDb ratings for each movie

# Loop through all movie soup objects using their index
for idx in range(len(movie_links)):
    try:
        # Retrieve the corresponding BeautifulSoup object from the stored globals
        soup = globals()[f"soup_movie_{idx}"]
        imdb_rating = 'N/A'  # Default value in case rating is not found

        # Look for the <img> tag with alt="IMDB", which usually appears next to the rating
        imdb_img = soup.find('img', {'alt': 'IMDB'})

        if imdb_img:
            # Find the next <div> with class "nowrap" after the IMDb logo
            # This div typically contains the IMDb rating text (e.g., "7.8")
            rating_div = imdb_img.find_next('div', class_='nowrap')

            if rating_div:
                # Use regular expression to search for a floating-point number like 7.8
                match = re.search(r'\d\.\d+', rating_div.get_text())
                # If match is found, extract the rating, else keep it as 'N/A'
                imdb_rating = match.group() if match else 'N/A'

        # Append the extracted or default rating to the imdb_ratings list
        imdb_ratings.append(imdb_rating)

    except Exception as e:
        # On any exception (e.g., tag not found, regex fails), append 'N/A'
        imdb_ratings.append('N/A')
        print(f"IMDb error at index {idx} →", e)  # Print error for debugging

# Display the first 5 extracted IMDb ratings to verify correctness
print("IMDb Ratings Extracted:", imdb_ratings[:5])
````

## **Scrapping Runtime/Duration**

````python
# Write Your Code here
# STEP 2F: Extract Movie Duration (Runtime) for each movie page
durations = []  # Initialize an empty list to store the runtime of each movie

# Loop through all movies by their index (corresponding to soup_movie_*)
for idx in range(len(movie_links)):
    try:
        # Retrieve the BeautifulSoup object for the current movie
        soup = globals()[f"soup_movie_{idx}"]
        duration = 'N/A'  # Default value in case runtime is not found

        # Find all <h3> tags with the specific class that labels metadata fields like Runtime, Genre, etc.
        h3_tags = soup.find_all('h3', class_='poster-detail-infos__subheading')

        # Loop through the found <h3> tags to locate the one labeled "Runtime"
        for h3 in h3_tags:
            if 'Runtime' in h3.get_text(strip=True):
                # The actual duration value is usually in a <div> just after the <h3> tag
                duration_div = h3.find_next_sibling('div', class_='poster-detail-infos__value')

                if duration_div:
                    # Extract and clean the text from the duration <div> (e.g., "1h 45m")
                    duration = duration_div.get_text(strip=True)
                break  # Exit the loop once "Runtime" is found and processed

        # Append the extracted runtime (or 'N/A') to the list
        durations.append(duration)

    except Exception as e:
        # In case of any error (e.g., tag missing or parsing issue), append 'N/A'
        durations.append('N/A')
        print(f"Duration extraction error at index {idx} →", e)  # Print error message for debugging

# Show the first 5 extracted durations to verify the output
print("Extracted Durations:", durations[:5])
````

## **Scrapping Age Rating**

````python
# Write Your Code here
# STEP 2G (Final): Extract Age Ratings using a heading-based pattern
age_ratings = []  # Initialize a list to store age ratings for each movie

# Loop through each movie index to process its corresponding BeautifulSoup object
for idx in range(len(movie_links)):
    try:
        # Retrieve the parsed HTML (soup) for the current movie
        soup = globals()[f"soup_movie_{idx}"]
        age = 'N/A'  # Default value in case age rating is not found

        # Find all <h3> tags that serve as section headings on the page (like Genre, Runtime, Age rating, etc.)
        h3_tags = soup.find_all('h3', class_='poster-detail-infos__subheading')

        # Loop through each heading tag to locate the one labeled "Age rating"
        for h3 in h3_tags:
            if 'Age rating' in h3.get_text(strip=True):  # Match text even if there are extra spaces
                # The age rating value is typically in the <div> immediately following the <h3>
                age_div = h3.find_next_sibling('div', class_='poster-detail-infos__value')
                if age_div:
                    # Extract the cleaned text (e.g., "U/A 13+", "PG-13", etc.)
                    age = age_div.get_text(strip=True)
                break  # Exit the loop once the "Age rating" section is found and processed

        # Append the result (found or 'N/A') to the age_ratings list
        age_ratings.append(age)

    except Exception as e:
        # If an error occurs (tag not found, parsing issue, etc.), store 'N/A' as a placeholder
        age_ratings.append('N/A')
        print(f"Age rating extraction error at index {idx} →", e)  # Print error for debugging

# Show a preview of the first 5 extracted age ratings to confirm success
print("Extracted Age Ratings:", age_ratings[:5])
````

## **Fetching Production Countries Details**

````python
# Write Your Code here
# STEP 2H: Extract Production Countries for each movie
countries = []  # Initialize an empty list to store the country of origin for each movie

# Loop over all movie indices (aligned with soup_movie_* objects)
for idx in range(len(movie_links)):
    try:
        # Retrieve the BeautifulSoup object for the current movie
        soup = globals()[f"soup_movie_{idx}"]
        country = 'N/A'  # Default value in case the production country is not found

        # Find all <h3> tags that act as metadata section headers on the page
        h3_tags = soup.find_all('h3', class_='poster-detail-infos__subheading')

        # Loop through all heading tags to find the one labeled "Production country"
        for h3 in h3_tags:
            if 'Production country' in h3.get_text(strip=True):
                # Once found, get the next <div> sibling, which contains the actual country info
                country_div = h3.find_next_sibling('div', class_='poster-detail-infos__value')

                if country_div:
                    # Extract the text content of the country <div>, and clean it
                    country_text = country_div.get_text(strip=True)
                    country = country_text if country_text else 'N/A'  # Set to 'N/A' if it's empty
                break  # Exit loop after finding the production country section

        # Append the extracted (or default) country value to the list
        countries.append(country)

    except Exception as e:
        # If an error occurs (e.g., tag missing, parsing issue), append 'N/A'
        countries.append('N/A')
        print(f"Country extraction error at index {idx} →", e)  # Print the error for debugging

# Display the first 5 extracted production countries to confirm successful scraping
print("Extracted Production Countries:", countries[:5])
````

## **Fetching Streaming Service Details**

````python
# Write Your Code here
# STEP 2I: Extract Streaming Services using <img title="...">
platforms = []  # Initialize an empty list to store available streaming platforms for each movie

# Loop through each movie index to access its corresponding soup object
for idx in range(len(movie_links)):
    try:
        # Retrieve the BeautifulSoup object for the current movie
        soup = globals()[f"soup_movie_{idx}"]
        streaming = 'N/A'  # Default value if no platforms are found

        # Find all <img> tags where:
        # - the tag has a 'title' attribute (used by JustWatch to store platform names like "Netflix")
        # - the class contains "provider-icon" (used to identify streaming service icons)
        provider_imgs = soup.find_all(
            'img',
            {
                'title': True,
                'class': re.compile(r'provider-icon')  # Uses regex to match classes containing 'provider-icon'
            }
        )

        # Extract the 'title' attribute (platform name) from each <img>, excluding "JustWatch" itself
        services = [
            img['title'].strip()
            for img in provider_imgs
            if img.get('title') and img['title'].strip().lower() != 'justwatch'
        ]

        # If services are found:
        # - Remove duplicates using set
        # - Sort alphabetically
        # - Join into a comma-separated string
        streaming = ", ".join(sorted(set(services))) if services else 'N/A'

        # Append the cleaned string (or 'N/A') to the final result list
        platforms.append(streaming)

    except Exception as e:
        # In case of any error (missing tag, network issue, etc.), append 'N/A' as a fallback
        platforms.append('N/A')
        print(f"Streaming service extraction error at index {idx} →", e)  # Print error for debugging

# Show a sample of the extracted streaming services to verify success
print("Extracted Streaming Services:", platforms[:5])
````

````python
# STEP 5: Full Data Quality Cleanup and save movie data cleaned as csv.

import pandas as pd         # For DataFrame operations
import numpy as np          # For handling missing values (NaN)
import re                   # For regular expression operations used in string cleaning

# --- Clean Genre column ---
def clean_genre(genre_str):
    # Handle missing or placeholder genre values
    if pd.isna(genre_str) or genre_str.strip().lower() in ["n/a", ""]:
        return "N/A"

    # Strip whitespace and remove any leading commas (e.g., ", Action" → "Action")
    cleaned = genre_str.strip().lstrip(',').strip()

    # Split the string into individual genres and clean each genre name
    parts = [g.strip() for g in cleaned.split(',')]

    # Remove duplicates and empty strings, then sort alphabetically
    unique_parts = sorted(set([p for p in parts if p]))

    # Join cleaned genres back into a comma-separated string
    return ', '.join(unique_parts) if unique_parts else "N/A"

# --- Clean title text and remove (year) from title ---
def clean_title(title):
    if isinstance(title, str):
        # Remove extra whitespace and non-breaking spaces
        title = title.strip().replace('\xa0', ' ')

        # Remove the year in parentheses at the end of the title (e.g., "Inception (2010)" → "Inception")
        return re.sub(r'\s*\(\d{4}\)$', '', title).strip()
    return title  # Return as-is if not a string (e.g., None)

# --- Clean all general text fields ---
def clean_text(text):
    # Removes non-breaking spaces and strips whitespace for basic string fields
    return text.strip().replace('\xa0', ' ') if isinstance(text, str) else text

# --- Apply all cleaning functions to the DataFrame columns ---

# Clean movie titles: trim, fix spacing, and remove trailing year
movies_df["Title"] = movies_df["Title"].apply(clean_title)

# Clean genre strings: remove duplicates, fix formatting
movies_df["Genre"] = movies_df["Genre"].apply(clean_genre)

# Replace "N/A" with NaN (missing) for numerical or structured fields
movies_df["IMDb Rating"] = movies_df["IMDb Rating"].replace("N/A", np.nan)
movies_df["Duration"] = movies_df["Duration"].replace("N/A", np.nan)
movies_df["Age Rating"] = movies_df["Age Rating"].replace("N/A", np.nan)

# Clean and standardize text fields (country and streaming platforms)
movies_df["Production Country"] = movies_df["Production Country"].apply(clean_text)
movies_df["Streaming Services"] = movies_df["Streaming Services"].apply(clean_text)

# --- Optional: Drop duplicate movies (if any) based on the URL ---
# Ensures only one entry per unique movie link
movies_df.drop_duplicates(subset=["URL"], keep='first', inplace=True)

# --- Reset the DataFrame index after cleaning ---
movies_df.reset_index(drop=True, inplace=True)

# Preview a few rows of the cleaned movie dataset
print("Cleaned Movie Data Sample:")
print(movies_df.head())

# STEP 6: Save cleaned movie data to CSV
# Exports the cleaned DataFrame to a CSV file named 'justwatch_movies_cleaned.csv'
movies_df.to_csv("justwatch_movies_cleaned.csv", index=False)

print("Cleaned movie data saved successfully as 'justwatch_movies_cleaned.csv'")
````

## **Analyzing Top Genres**

````python
# Write Your Code here
# --- Define a function to extract individual genres from the 'Genre' column of a DataFrame ---
def extract_genres(df):
    genre_list = []  # This list will collect all valid genres from the dataset

    # Iterate through non-null values in the 'Genre' column
    for genres in df['Genre'].dropna():

        # Ensure the value is a string and not marked as "N/A"
        if isinstance(genres, str) and genres.strip().lower() != "n/a":

            # Split multiple genres (comma-separated), e.g. "Action, Drama"
            for g in genres.split(','):
                g_clean = g.strip()  # Remove leading/trailing whitespace

                if g_clean:          # Only append if genre string is not empty
                    genre_list.append(g_clean)

    return genre_list  # Return list of all genres found

# --- Step 1: Extract genre lists separately from movie and TV datasets ---

movie_genres = extract_genres(movies_df)  # Extract genres from movies_df
tv_genres = extract_genres(tv_df)         # Extract genres from tv_df

# Combine both lists to get a full view across all content
all_genres = movie_genres + tv_genres

# --- Step 2: Use Counter to find the top 5 most common genres in each category ---

from collections import Counter  # (If not already imported)

movie_top_genres = Counter(movie_genres).most_common(5)      # Top 5 movie genres
tv_top_genres = Counter(tv_genres).most_common(5)            # Top 5 TV show genres
combined_top_genres = Counter(all_genres).most_common(5)     # Top 5 across both

# --- Step 3: Display the results ---

print("\n Top 5 Movie Genres:")
for genre, count in movie_top_genres:
    print(f"  {genre}: {count}")  # Print genre and how many times it appeared

print("\n Top 5 TV Show Genres:")
for genre, count in tv_top_genres:
    print(f"  {genre}: {count}")

print("\n Top 5 Combined Genres:")
for genre, count in combined_top_genres:
    print(f"  {genre}: {count}")
````

## **Finding Predominant Streaming Service**

````python
# Write Your Code here
# --- Function to extract individual streaming platforms from a DataFrame column ---
def extract_services(df):
    services = []  # This list will store all valid streaming service names

    # Loop through all non-null entries in the 'Streaming Services' column
    for entry in df["Streaming Services"].dropna():

        # Ensure the entry is a string and not marked as "N/A"
        if isinstance(entry, str) and entry.strip().lower() != "n/a":

            # Split multiple services (comma-separated), e.g. "Netflix, Prime Video"
            for s in entry.split(','):
                s_clean = s.strip()  # Clean whitespace

                if s_clean:  # Only add non-empty values
                    services.append(s_clean)

    return services  # Return the full list of streaming service names

# --- Step 1: Extract all services separately from movies and TV show data ---

movie_services = extract_services(movies_df)  # Services used by movies
tv_services = extract_services(tv_df)          # Services used by TV shows

# Combine both lists into one to analyze overall usage
all_services = movie_services + tv_services

# --- Step 2: Use Counter to identify the most common (top) streaming platform ---

from collections import Counter  # (Assumed already imported earlier)

movie_top_service = Counter(movie_services).most_common(1)       # Top service in movies
tv_top_service = Counter(tv_services).most_common(1)             # Top service in TV shows
combined_top_service = Counter(all_services).most_common(1)      # Top overall service

# --- Step 3: Print the results clearly ---

print("\n Top Streaming Service - Movies:")
# Use conditional check to avoid error if no data
print(f"  {movie_top_service[0][0]} : {movie_top_service[0][1]} titles" if movie_top_service else "  No valid data")

print("\n Top Streaming Service - TV Shows:")
print(f"  {tv_top_service[0][0]} : {tv_top_service[0][1]} titles" if tv_top_service else "  No valid data")

print("\n Top Streaming Service - Combined:")
print(f"  {combined_top_service[0][0]} : {combined_top_service[0][1]} titles" if combined_top_service else "  No valid data")

````
